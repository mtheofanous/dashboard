```python
# Function to add random time
def add_random_time(row):
    if row['estado_primer_movimiento'] == 'Volver a Contactar':
        if random.random() < 0.6:   
            # Add random hours between 1 and 24
            random_hours = np.random.randint(1, 25)
        else:
            # Add random hours between 25 and 48
            random_hours = np.random.randint(3, 49)
        return row['fecha_primer_movimiento'] + timedelta(hours=random_hours)
    else:
        # Add random minutes between 1 and 24
        if random.random() < 0.6:
            random_minutes = np.random.randint(1, 25)
        else:
            random_minutes = np.random.randint(10, 91)
        return row['fecha_primer_movimiento'] + timedelta(minutes=random_minutes)

# Apply function to create the new column
df['fecha_primer_contacto'] = df.apply(add_random_time, axis=1)

```

```python
# horas_primer_movimiento

def add_horas(row):
    if row['dias_entre_alta_y_primer_movimiento'] == 0:
        return 'Contacto el mismo dia'
    elif row['horas_entre_alta_y_primer_movimiento'] <= 24:
        return 'Contancto durante primeras 24 horas'
    elif 24 < row['horas_entre_alta_y_primer_movimiento'] <= 48:
        return 'Contacto durante primeras 48 horas'
    elif 48 < row['horas_entre_alta_y_primer_movimiento'] <= 72:
        return 'Contacto durante primeras 72 horas'
    else:
        return 'Contacto pasadas 72 horas'
    
df['horas_primer_movimiento'] = df.apply(add_horas, axis=1)
```

```python
# dias_entre_alta_y_primer_contacto
df['dias_entre_alta_y_primer_contacto'] = (df['fecha_primer_contacto'] - df['fecha_alta']).dt.days

# horas_entre_alta_y_primer_contacto

df['horas_entre_alta_y_primer_contacto'] = [int((f - a).total_seconds() / 3600) for a, f in zip(df['fecha_alta'], df['fecha_primer_contacto'])]
```

```python
# horas_primer_contacto

def add_horas_contacto(row):
    if row['dias_entre_alta_y_primer_contacto'] == 0:
        return 'Contacto el mismo dia'
    elif row['horas_entre_alta_y_primer_contacto'] <= 24:
        return 'Contancto durante primeras 24 horas'
    elif 24 < row['horas_entre_alta_y_primer_contacto'] <= 48:
        return 'Contacto durante primeras 48 horas'
    elif 48 < row['horas_entre_alta_y_primer_contacto'] <= 72:
        return 'Contacto durante primeras 72 horas'
    else:
        return 'Contacto pasadas 72 horas'
    
df['horas_primer_contacto'] = df.apply(add_horas_contacto, axis=1)
```

```python
# fecha_ultimo_estado

# Function to add random time
def add_random_time(row):
    if row['estado_primer_movimiento'] == 'Volver a Contactar':
        # Add random hours between 3 and 48
        random_hours = np.random.randint(3, 49)
        return row['fecha_primer_contacto'] + timedelta(hours=random_hours)
    else:
        # Add random minutes between 10 and 90
        random_minutes = np.random.randint(10, 49)
        return row['fecha_primer_contacto'] + timedelta(minutes=random_minutes)
    
# Apply function to create the new column
df['fecha_ultimo_estado'] = df.apply(add_random_time, axis=1)

```

```python
# estado_ultimo_estado

estados = ['Cliente no Localizado', 'Incidencia', 'Pendiente servicio', 'Valoraci√≥n Medica Pendiente Internos', 'Valorado Interno', 'Volver a Contactar']
estados_finales = ['Cliente no Localizado', 'Incidencia', 'Cerrado']

def estados(row):
    if row['estado_primer_movimiento'] == 'Volver a Contactar':
        return random.choices(estados_finales, weights=[0.1, 0.1, 0.8])[0]
    elif row['estado_primer_movimiento'] in ['Valoraci√≥n Medica Pendiente Internos', 'Valorado Interno', 'Pendiente servicio']:
        return 'Cerrado'
    else:
        return row['estado_primer_movimiento']
    
df['estado'] = df.apply(estados, axis=1)
    
```
